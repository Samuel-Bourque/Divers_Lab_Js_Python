Le fichier est structur√© comme suit:

| champ        | longueur et autres informations
|--------------|---------------------------------------------
| race         | 6 caract√®res (encodage ASCII)
| classe       | 6 caract√®res (encodage ASCII)
| force        | nombre sur un octet compris entre 6 et 18
| dext√©rit√©    | nombre sur un octet compris entre 6 et 18
| constitution | nombre sur un octet compris entre 6 et 18
| intelligence | nombre sur un octet compris entre 6 et 18
| sagesse      | nombre sur un octet compris entre 6 et 18
| charisme     | nombre sur un octet compris entre 6 et 18

Le fichier n'est compos√© que d'un seul enregistrement.  

Votre mandat est d'afficher les informations r√©cup√©r√©es du fichier et les afficher.

**R√©sultats attendus (note: les valeurs sont remplac√©es par des points d'interrogation)**:

````python
import struct

with open('perso.dat','rb') as f:
    infos = struct.unpack('6s6sbbbbbb',f.read())
    print('Race        :',infos[0].decode('ascii')) 
    print('Classe      :',infos[1].decode('ascii'))
    print('Force       :',infos[2])
    print('Dext√©rit√©   :',infos[3])
    print('Constitution:',infos[4])
    print('Intelligence:',infos[5])
    print('Sagesse     :',infos[6])
    print('Charisme    :',infos[7])
````

Un fichier poss√®de un **nombre entier non-sign√© sur 2 octets** et un **nombre entier sign√© sur 4 octets**.

En lisant le fichier (une ou deux fois), vous devez d√©terminer l'ordre des octets et afficher:
- l'ordre des octets (little- ou big-endian)
- la valeur du second nombre

````python
!xxd mystere.dat
````

````python
import struct
with open('mystere.dat','rb') as f:
    enreg = struct.unpack('<hl', f.read())    
    print('Ordre: little-endian')
    print('Second nombre:', enreg[1])
````
